```python
import numpy as np
import struct
import matplotlib.pyplot as plt

def load_particles_bin(filename):
    with open(filename, "rb") as f:
        n = struct.unpack("i", f.read(4))[0]
        data = np.fromfile(f, dtype=np.float32).reshape(n, 7)
    return data

particles = load_particles_bin("particles.bin")
print("Loaded", len(particles), "particles.")
print("First particle:", particles[0])
```

```python
x = particles[:, 1]
y = particles[:, 2]

plt.scatter(x, y, s=1)
plt.show
```
