```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

outfolder = 'tree_output/'
df = pd.read_csv(outfolder + "summary_results.csv")
name = "Tree"
# ----------- Plot 1: Error vs Theta -----------

plt.loglog(df["theta"], df["Ex_error"], marker='x', label=name+" Fx_error")

outfolder = 'fmm_output/'
df = pd.read_csv(outfolder + "summary_results.csv")
name = "FMM"
# ----------- Plot 1: Error vs Theta -----------

plt.loglog(df["theta"], df["Ex_error"], marker='x', label=name+" Fx_error")



plt.xlabel("Theta")
plt.ylabel("Relative Error")
plt.title("Force Error vs Theta")
plt.legend()
# plt.grid(True)
plt.tight_layout()
plt.savefig(outfolder + "error_vs_theta.png")
plt.show()

```

```python
time_unit = 1000
# ----------- Plot 2: Time vs Theta -----------
plt.figure(figsize=(8, 6))

outfolder = 'tree_output/'
df = pd.read_csv(outfolder + "summary_results.csv")
name = "Tree"

plt.loglog(df["theta"], df["Total execution time"]/time_unit, marker='x', label=name+" Total execution time")

outfolder = 'fmm_output/'
df = pd.read_csv(outfolder + "summary_results.csv")
name = "FMM"
plt.loglog(df["theta"], df["Total execution time"]/time_unit, marker='x', label=name+" Total execution time")
plt.loglog(df["theta"], df["theta"]**(-3), linestyle='--', label=name+"N^-3")


plt.xlabel("Theta")
plt.ylabel("Time (s)")
plt.title("Execution Time vs Theta")
plt.legend()
# plt.grid(True)
plt.tight_layout()
plt.savefig(outfolder + "time_vs_theta.png")
plt.show()
```
